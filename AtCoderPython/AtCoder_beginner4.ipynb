```python
total_num = int(input())

nums = list(map(int, input().split()))

count = 0

odd_doesnt_exist = True

while odd_doesnt_exist:
    for i in range(total_num):
        if nums[i]%2 == 1:
            odd_doesnt_exist = False
            break
        nums[i] = nums[i]/2
        print(nums)
        if i == total_num-1:
            count = count + 1


print(count)


```
